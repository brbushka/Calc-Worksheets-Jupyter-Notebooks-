```python
%%manim TangentDrawer -qm -v WARNING
class TangentDrawer(Scene):
    def construct(self):
        axes = Axes(
            x_range=[-10, 10, 1],
            x_length=9,
            y_range=[0, 20, 5],
            y_length=6,
            axis_config={"include_numbers": True, "include_tip": True},
        )
        axes.to_edge(DL)
        axes.set_color(GREY)

        axes_labels = axes.get_axis_labels(x_label_tex="x", y_label_tex="y")

        graph = axes.get_graph(
            lambda x: np.sin(x), x_range=[-10, 10], color=BLUE
        )

        graph_der = axes.get_graph(
            lambda x: np.cos(x), x_range=[-10, 10], color=RED
        )

        x = ValueTracker(-5)

        pt = always_redraw(lambda: Dot().move_to(axes.c2p(x.get_value(), graph.underlying_function(x.get_value()))))
        der_pt = always_redraw(lambda: Dot().move_to(axes.c2p(x.get_value(), graph_der.underlying_function(x.get_value()))))
        path = TracedPath(der_pt.get_center, stroke_color=RED)


        def get_line(x):
          m = graph_der.underlying_function(x)
          b = graph.underlying_function(0)
          return Line(start = axes.c2p(x - 1, m * (x - 1) + b - x * m + graph.underlying_function(x)),
                      end = axes.c2p(x + 1, m * (x + 1) + b - x * m + graph.underlying_function(x)))
        line = always_redraw(lambda: get_line(x.get_value()))

        self.add(axes, axes_labels, graph, pt, der_pt, path, line)
        self.play(x.animate.set_value(4), run_time=5)
        self.wait()
```
